```python
# Example: Predict if a student will pass (1) or fail (0)
# Step 1: Import Libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

# Step 2: Create a Simple Manual Dataset
data = {
    'StudyHours': [2, 4, 6, 8, 10, 1, 3, 5, 7, 9],
    'Attendance': [50, 60, 65, 80, 90, 30, 55, 70, 85, 95],
    'Passed':     [0, 0, 1, 1, 1, 0, 0, 1, 1, 1]
}

df = pd.DataFrame(data)
print("âœ… Simple Dataset:")
print(df)

# Step 3: Split Data
X = df[['StudyHours', 'Attendance']]
y = df['Passed']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Build AdaBoost Model
base = DecisionTreeClassifier(max_depth=1)
ada = AdaBoostClassifier(estimator=base, n_estimators=20, learning_rate=1.0, random_state=42)

# Step 5: Train Model
ada.fit(X_train, y_train)

# Step 6: Predict
y_pred = ada.predict(X_test)

# Step 7: Evaluate
print("\nâœ… Accuracy:", round(accuracy_score(y_test, y_pred)*100, 2), "%")
print("\nðŸ“Š Classification Report:\n", classification_report(y_test, y_pred))
```
